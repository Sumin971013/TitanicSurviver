```python
from sklearn.model_selection import train_test_split
```

```python
train_data=train.drop('Survived',axis=1)
target_data=train['Survived']
```

```python
x_train,x_valid,y_train,y_valid=train_test_split(train_data,target_data)
```

```python
print(train_data.shape,x_train.shape,x_valid.shape)
```

```python
#scikit learn의 decision classifier로 학습하기
from sklearn.tree import DecisionTreeClassifier
```

```python
#tree 객체 생성
tree=DecisionTreeClassifier()
```

```python
tree.fit(x_train,y_train)
print('traing set accuracy:',tree.score(x_train,y_train))
print('valid set accuracy:',tree.score(x_valid,y_valid))
```

```python
prediction=tree.predict(test)
prediction
```

```python
#결과 csv로 만들기
result=pd.DataFrame({
    'PassengerId':test1['PassengerId'],
    'survived':prediction
})
#891부터의 test 데이터의 승객의 사망 여부를 측정한 결과를 csv파일로 생성
result.to_csv('titanic_Result.csv',index=False)

```
